```python
from model.databasetemp import get_temps, get_timestamps
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.io import export_png, curdoc
from bokeh.themes import Theme
curdoc().theme = Theme(filename="themes/redTheme.yml")
```

```python
yget = get_temps()
xget = get_timestamps()

x = []
y = []
z = []

for i in xget:
    x.append(i[0])

for i in yget:
    z.append(float(i[0])+1)

for i in yget:
    y.append(float(i[0]))
    
xs = [[x],[y]]
ys = [[x],[z]]
```
